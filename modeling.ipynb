## Section 5 - Merge data
In order to include all of the data within a predictive model, merging them together into 1 dataframe is a must.

Recall the problem of statement:
```
Can we accurately predict the stock levels of products, based on sales data and sensor data, on an hourly basis in order to more intelligently procure products from our suppliers.
```
The client indicates that they want the model to predict on an hourly basis. Looking at the data model, there's only one column that can be used to merge the 3 datasets together & that is `timestamp`. Thus, first transform the `timestamp` column in all 3 datasets to be based on the hour of the day, then merge the datasets together.

````python
sales_df.head()
````

````python
from datetime import datetime
def convert_timestamp_to_hourly(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  new_ts = dummy[column].tolist()
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in new_ts]
  dummy[column] = new_ts
  return dummy
````

````python
sales_df = convert_timestamp_to_hourly(sales_df, 'timestamp')
sales_df.head()
````

````python
stock_df = convert_timestamp_to_hourly(stock_df, 'timestamp')
stock_df.head()
````

````python
temp_df = convert_timestamp_to_hourly(temp_df, 'timestamp')
temp_df.head()
````

Now, for all `timestamp` columns have had the minutes and seconds reduced to `00`. Next thing is to aggregate the datasets in order to combine rows which have the same value for `timestamp`.
For the `sales` data, group the data by `timestamp` but also by `product_id`. When aggregating, choose columns to aggregate by the grouping. For now, aggregate quantity.

````python
sales_agg = sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_agg.head()
````

From the aggregatted sales data, each row represents a unique combination of hour during which the sales took place from that weeks worth of data and the product_id. This is summed quantity and taking the mean average of the unit_price. For the stock data, group it the same way and aggregate the `estimated_stock_pct`.

````python
stock_agg = stock_df.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_agg.head()
````

This shows the average stock percentage of each product at unique hours within the week of sample data.
Finally, for the temperature data, product_id does not exist in this table thus the need to group by timestamp and aggregate the `temperature`.

````python
temp_agg = temp_df.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_agg.head()
````

This gives the average temperature of the storage facility where the produce is stored in the warehouse by unique hours during the week. To merge the data, use the stock_agg table as base table, and merge the other 2 tables onto this.

````python
merged_df = stock_agg.merge(sales_agg, on=['timestamp', 'product_id'], how='left')
merged_df.head()
````

````python
merged_df = merged_df.merge(temp_agg, on='timestamp', how='left')
merged_df.head()
````

````python
merged_df.info()
````

There are some null values, they need to be treated before building a predictive model. The column that features some null values is quantity. Assume that if there is a null value for this column, it represents that there were 0 sales of this product within this hour. So, fill this columns null values with 0, however, verify this with the client in order to make sure that there are any assumptions made by filling these null values with 0.

````python
merged_df['quantity'] = merged_df['quantity'].fillna(0)
merged_df.info()
````

Combining some more features onto this table too, including `category` and `unit_price`.

````python
product_categories = sales_df[['product_id', 'category']]
product_categories = product_categories.drop_duplicates()
product_price = sales_df[['product_id', 'unit_price']]
product_price = product_price.drop_duplicates()
````

````python
merged_df = merged_df.merge(product_categories, on="product_id", how="left")
merged_df.head()
````

````python
merged_df = merged_df.merge(product_price, on="product_id", how="left")
merged_df.head()
````

````python
merged_df.info()
````

`Table with 2 extra features!`

##Section 6 - Feature engineering
Having cleaned and merged the data, now it must be transformed so that the columns are in a suitable format for a ML-Model. In other terms, every column must be numeric. There are some models that will accept categorical features, but for this case it will require numeric features.

First, engineer the timestamp column. In it's current form, it is not very useful for a machine learning modeL since it's a datetime datatype. Explode this column into day of week, day of month and hour to name a few.

````python
merged_df['timestamp_day_of_month'] = merged_df['timestamp'].dt.day
merged_df['timestamp_day_of_week'] = merged_df['timestamp'].dt.dayofweek
merged_df['timestamp_hour'] = merged_df['timestamp'].dt.hour
merged_df.drop(columns=['timestamp'], inplace=True)
merged_df.head()
````

The next column that can be engineered is the category column. In its current form it is categorical. Convert it into numeric by creating dummy variables from this categorical column - a dummy variable's a binary flag column (1's and 0's) that indicates whether a row fits a particular value of that column. For example, one can create a dummy column called category_pets, which will contain a 1 if that row indicates a product which was included within this category and a 0 if not.

````python
merged_df = pd.get_dummies(merged_df, columns=['category'])
merged_df.head()
````

````python
merged_df.info()
````

Looking at the table, there's 1 remaining column which is not numeric that is the `product_id`. Since each row represents a unique combination of `product_id` and `timestamp` by hour, and the `product_id` is simply an ID column, it will add no value by including it in the predictive model. Hence, remove it from the modeling process.

````python
merged_df.drop(columns=['product_id'], inplace=True)
merged_df.head()
````

This feature engineering by no means is/was exhaustive, but enough to give an example of the process followed when engineering the features of a dataset. In reality, this is an iterative task. Once a model is built, there's need to revist feature engineering in order to create new features to boost the predictive power of a ML-Model.

##Section 7 - Modelling
Now it is time to train a ML-Model. Uing a supervised ML-Model, and using `estimated_stock_pct` as the target variable, since the problem statement was focused on being able to predict the stock levels of products on an hourly basis.

Whilst training the  ML-Model, use cross-validation, which is a technique where one holds back a portion of the dataset for testing in order to compute how well the trained ML-Model is able to predict the target variable.

Finally, to ensure that the trained ML-Model is able to perform robustly, there's a want to test it several times on random samples of data, not just once. Hence, use a K-fold strategy to train the ML-Model on K (K is an integer to be decided) random samples of the data.

First, create a target variable `y` and independent variables `X`

````python
X = merged_df.drop(columns=['estimated_stock_pct'])
y = merged_df['estimated_stock_pct']
print(X.shape)
print(y.shape)
````

This shows that we have 29 predictor variables that we will train the ML-Model on and 10845 rows of data. Now, define how many folds will complete during training, and how much of the dataset to assign to training, leaving the rest for test. Typically, a gap of at least 20-30% of the data should be left for testing.

````python
K = 10
split = 0.75
````

Here, use a RandomForestRegressor model which's an instance of a Random Forest. These are powerful tree based ensemble algorithms and are particularly good because their results are very interpretable. Use a regression algorithm here because of prediction of a continuous numeric variable, that is, `estimated_stock_pct`.

Use `scikit-learn` package for the machine learning algorithm.

````python
!pip install scikit-learn
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
````

`Create a loop to train K-models with a 75/25% random split of the data each time between training and test samples`

````python
accuracy = []
for fold in range(0, K):
  # Instantiate algorithm
  model = RandomForestRegressor()
  scaler = StandardScaler()
  # Create training and test samples
  X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=split, random_state=42)
  # Scale X data to help the algorithm converge and also not to be greedy with large values
  scaler.fit(X_train)
  X_train = scaler.transform(X_train)
  X_test = scaler.transform(X_test)
  # Train the model
  trained_model = model.fit(X_train, y_train)
  # Generate predictions on test sample
  y_pred = trained_model.predict(X_test)
  # Compute accuracy, using mean absolute error
  mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
  accuracy.append(mae)
  print(f"Fold {fold + 1}: MAE = {mae:.3f}")
print(f"Average MAE: {(sum(accuracy) / len(accuracy)):.2f}")
````

The mean absolute error (MAE) is almost exactly the same each time - this is a good sign, it shows that the performance of the model is consistent across different random samples of the data, which is what is expected - it shows a robust nature.

The MAE was chosen as a performance metric because it describes how closely the ML-Model was able to predict the exact value of `estimated_stock_pct`.

Even though the model is predicting robustly, this value for MAE is not so good, since the average value of the target variable is around 0.51, meaning that the accuracy as a percentage was around 50%. In an ideal world, the MAE should be as low as possible. This is where the iterative process of ML comes in.
